```python
import torch
from torch.utils.data import Dataset, DataLoader
import cv2
import numpy as np

# Sprint 1'de oluşturduğumuz ve doğruladığımız 'train_df' değişkenini kullanıyoruz.
# Not: Eğer oturum kapanırsa, kaydettiğin CSV'yi train_df = pd.read_csv("dosya_yolun.csv") ile okumalısın.

class MicrobeDataset(Dataset):
    def __init__(self, dataframe, target_size=(224, 224)):
        self.df = dataframe
        self.target_size = target_size

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        # 1. Yol Bilgilerini Al
        row = self.df.iloc[idx]
        img_path = row['image_path']
        mask_path = row['mask_path']

        # 2. Görüntüyü Oku ve Ön İşle (Şeyma'nın standardı: 224x224)
        image = cv2.imread(img_path)
        image = cv2.resize(image, self.target_size)
        image = image.astype(np.float32) / 255.0  # Normalizasyon

        # PyTorch formatına çevir (H, W, C -> C, H, W)
        image = torch.from_numpy(image).permute(2, 0, 1)

        # 3. Maskeyi Oku ve Ön İşle (Aybüke'nin U-Net eğitimi için)
        mask = cv2.imread(mask_path, cv2.IMREAD_UNCHANGED)
        mask = cv2.resize(mask, self.target_size, interpolation=cv2.INTER_NEAREST)

        # Maskeyi binary (0-1) formatına çevir ve Tensor yap
        mask = (mask > 0).astype(np.float32)
        mask = torch.from_numpy(mask).unsqueeze(0) # Kanal boyutu ekle (1, H, W)

        return image, mask

# Dataset ve DataLoader oluşturma
# train_df senin son kodunda kullandığın, 1568 satırlık ana veri çerçeven olmalı
full_dataset = MicrobeDataset(train_df)
train_loader = DataLoader(full_dataset, batch_size=16, shuffle=True)

# Test amaçlı bir batch çekelim
images, masks = next(iter(train_loader))
print(f"Batch Görüntü Boyutu: {images.shape}") # [16, 3, 224, 224] bekliyoruz
print(f"Batch Maske Boyutu: {masks.shape}")     # [16, 1, 224, 224] bekliyoruz
```

```python
import torch.nn as nn
import torch.nn.functional as F

class UNet(nn.Module):
    def __init__(self):
        super(UNet, self).__init__()

        # Encoder (Küçültücü Yol)
        self.enc1 = self.conv_block(3, 64)
        self.enc2 = self.conv_block(64, 128)
        self.enc3 = self.conv_block(128, 256)
        self.enc4 = self.conv_block(256, 512)

        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)

        # Bottleneck (Darboğaz)
        self.bottleneck = self.conv_block(512, 1024)

        # Decoder (Genişletici Yol)
        self.up4 = nn.ConvTranspose2d(1024, 512, kernel_size=2, stride=2)
        self.dec4 = self.conv_block(1024, 512)
        self.up3 = nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2)
        self.dec3 = self.conv_block(512, 256)
        self.up2 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)
        self.dec2 = self.conv_block(256, 128)
        self.up1 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)
        self.dec1 = self.conv_block(128, 64)

        # Final Katman (Piksel bazlı tahmin)
        self.final_conv = nn.Conv2d(64, 1, kernel_size=1)

    def conv_block(self, in_ch, out_ch):
        return nn.Sequential(
            nn.Conv2d(in_ch, out_ch, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        # Encoder
        e1 = self.enc1(x)
        e2 = self.enc2(self.pool(e1))
        e3 = self.enc3(self.pool(e2))
        e4 = self.enc4(self.pool(e3))

        # Bottleneck
        b = self.bottleneck(self.pool(e4))

        # Decoder (Skip Connections dahil)
        d4 = self.up4(b)
        d4 = torch.cat((d4, e4), dim=1) # E4 ile birleştir
        d4 = self.dec4(d4)

        d3 = self.up3(d4)
        d3 = torch.cat((d3, e3), dim=1) # E3 ile birleştir
        d3 = self.dec3(d3)

        d2 = self.up2(d3)
        d2 = torch.cat((d2, e2), dim=1) # E2 ile birleştir
        d2 = self.dec2(d2)

        d1 = self.up1(d2)
        d1 = torch.cat((d1, e1), dim=1) # E1 ile birleştir
        d1 = self.dec1(d1)

        return torch.sigmoid(self.final_conv(d1))

# Modeli oluşturup GPU'ya gönderelim
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = UNet().to(device)
print(f"U-Net Modeli Oluşturuldu. Cihaz: {device}")
```

U-Net Eğitim Döngüsünü (Training Loop) başlatıyoruz. Bu aşamada model, görüntülere bakıp maskeleri tahmin etmeye çalışacak ve yaptığı hatalardan ders çıkaracak.

```python
import torch
import torch.nn as nn
import torch.optim as optim
import time

# 1. Dice Loss Fonksiyonu (Segmentasyon için kritik)
class DiceLoss(nn.Module):
    def __init__(self):
        super(DiceLoss, self).__init__()

    def forward(self, inputs, targets, smooth=1):
        inputs = inputs.view(-1)
        targets = targets.view(-1)

        intersection = (inputs * targets).sum()
        dice = (2.*intersection + smooth)/(inputs.sum() + targets.sum() + smooth)

        return 1 - dice

# 2. IoU Metriği (Başarı Kriterimiz: IoU >= 0.75)
def calculate_iou(pred, target, threshold=0.5):
    pred = (pred > threshold).float()
    intersection = (pred * target).sum()
    union = pred.sum() + target.sum() - intersection
    if union == 0:
        return 1.0
    return intersection / union

# Modeli hazırlayalım
criterion_dice = DiceLoss()
criterion_bce = nn.BCELoss() # Stabilite için BCE ile hibrit kullanacağız
optimizer = optim.Adam(model.parameters(), lr=1e-4)

num_epochs = 15
print(f"🚀 Eğitim Başlıyor... Hedef: IoU >= 0.75")

for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    total_iou = 0

    for images, masks in train_loader:
        images, masks = images.to(device), masks.to(device)

        optimizer.zero_grad()
        outputs = model(images)

        # Hibrit Kayıp: Hem piksel doğruluğu (BCE) hem bölge örtüşmesi (Dice)
        loss = criterion_bce(outputs, masks) + criterion_dice(outputs, masks)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        total_iou += calculate_iou(outputs, masks).item()

    avg_loss = total_loss / len(train_loader)
    avg_iou = total_iou / len(train_loader)

    print(f"Epoch [{epoch+1}/{num_epochs}] | Loss: {avg_loss:.4f} | IoU: {avg_iou:.4f}")

print("\n✅ Eğitim Tamamlandı.")
```

Bu, U-Net modelinin bakterilerin sınırlarını, şekillerini ve konumlarını neredeyse bir uzman gibi (insan eliyle çizilmiş maskelere çok yakın) öğrendiğini gösteriyor.

2. Öğrenme Eğrisi (Loss Analizi)Başlangıç: $1.0629$Bitiş: $0.1774$Yorum: Kayıp (Loss) değeri her epoch'ta istikrarlı bir şekilde düşmüş. Bu, modelde "overfitting" (ezberleme) veya "divergence" (sapma) gibi teknik sorunlar olmadığını, sağlıklı bir öğrenme gerçekleştiğini kanıtlar.


**1. Adım: Veri Etiketleme ve Hazırlık**
U-Net'te sadece görüntülere bakıyorduk, şimdi ise görüntünün ait olduğu klasörden "antibiyotik tipini" (etiket/label) öğrenmemiz gerekiyor.

```python
import numpy as np
import cv2

def apply_mask_and_crop(image, mask, threshold=0.5):
    # 1. Maskeyi binary yap
    mask = (mask > threshold).astype(np.uint8)

    # 2. Maskeyi genişlet (3 channel yap)
    mask_3ch = np.repeat(mask, 3, axis=0).transpose(1,2,0)

    # 3. Maskeyi uygula
    masked_img = image * mask_3ch

    # 4. Bounding box bul (sadece bakteriyi crop etmek için)
    coords = np.column_stack(np.where(mask.squeeze() > 0))

    if len(coords) == 0:
        # mask boşsa fallback
        return image

    y_min, x_min = coords.min(axis=0)
    y_max, x_max = coords.max(axis=0)

    cropped = masked_img[:, y_min:y_max, x_min:x_max]

    # 5. Resize (DenseNet için)
    cropped = cv2.resize(cropped.transpose(1,2,0), (224,224))
    cropped = cropped.transpose(2,0,1)

    return cropped
```

```python
class HybridDataset(Dataset):
    def __init__(self, dataframe, unet_model, device, transform=None):
        self.df = dataframe
        self.unet = unet_model
        self.device = device
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        # 1. Görüntüyü oku
        image = cv2.imread(row['image_path'])
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image = cv2.resize(image, (224,224))
        image = image.astype(np.float32) / 255.0

        image_tensor = torch.from_numpy(image).permute(2,0,1).unsqueeze(0).to(self.device)

        # 2. U-Net ile maske tahmini
        with torch.no_grad():
            pred_mask = self.unet(image_tensor)

        pred_mask = pred_mask.squeeze().cpu().numpy()

        # 3. Maskeyi uygula + crop
        #processed_img = apply_mask_and_crop(image_tensor.squeeze().cpu().numpy(), pred_mask)
        img_np = image_tensor.squeeze().cpu().numpy().transpose(1,2,0)
        processed_img = apply_mask_and_crop(img_np, pred_mask)

        # 4. Tensor’a çevir
        processed_img = torch.from_numpy(processed_img).float()

        # 5. Normalize (ImageNet)
        processed_img = (processed_img - 0.5) / 0.5

        label = int(row['label'])

        return processed_img, label
```
